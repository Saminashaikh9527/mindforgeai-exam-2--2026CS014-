** 1.Create or load the dataset and display the first rows**


```python
import pandas as pd
data = {
    "student_id":["2026CS001","2026CS002","2026IT001","2026IT002","2026CE001","2026CE002","2026CS003","2026IT003"],
    "name":["Aditi","Bhavana","Chaitra","Divya","Esha","Fatima","Gauri","Harshada"],
    "department":["Computer Engineering","Computer Engineering","Information Technology","Information Technology","Civil Engineering",
                  "Civil Engineering","Computer Engineering","Information Technology"],
    "attendance_pct":[92,64,78,55,88,71,96,43],
    "diary_entries":[12,6,9,4,10,8,14,2],
    "tasks_completed":[8,5,7,3,8,6,9,2],
    "assignment_score":[84,58,72,46,79,66,91,35],
    "lab_score":[88,61,76,52,82,69,94,41]
}

df = pd.DataFrame(data)

df.to_csv("week_2_3_exam_student_signals.csv",index=False)

df.head()
```

 **Explanation**

In this step, I created the student dataset and converted it into a DataFrame. Then I saved it as a CSV file and displayed the first five records.

**2.Check data types, missing values, and basic shape**

```python
print("Data Types")
print(df.dtypes)

print("\nMissing Values")
print(df.isnull().sum())

print("\nShape")
print(df.shape)
```

 **Explanation**

**3.Calculate the readiness score using attendance, diary entries, tasks completed, assignment score, and lab score.**

```python
def readiness_score(attendance, diary, tasks, assignment, lab):
    score = (
        attendance * 0.30 +
        diary * 2 +
        tasks * 3 +
        assignment * 0.25 +
        lab * 0.25
    )
    if score > 100:
        score = 100

    return round(score, 2)

df["readiness_score"] = df.apply(lambda x: readiness_score(  x["attendance_pct"], x["diary_entries"],x["tasks_completed"],
                                                             x["assignment_score"],x["lab_score"]), axis=1)

df[["name","readiness_score"]]
```

**Explanation**

In this step, I created a function to calculate the readiness score for each student. The score is calculated using attendance, diary entries, tasks completed, assignment score, and lab score. Then I added the readiness score as a new column in the dataset.

**4.Use loops or apply logic to classify each student into Support, Developing, Strong, or Excellent**

```python
performance = []

for score in df["readiness_score"]:
    if score < 45:
        performance.append("Support")
    elif score < 70:
        performance.append("Developing")
    elif score < 90:
        performance.append("Strong")
    else:
        performance.append("Excellent")

df["Performance_Band"] = performance

df[["name","readiness_score","Performance_Band"]]
```

**Explanation**

In this step, I used a loop to check the readiness score of each student. Based on the score, I classified each student as Support, Developing, Strong, or Excellent. Then I stored the result in a new column called Performance_Band.

**5. Generate at least three analytics outputs: department average, top performer, and students needing support.**

```python
print("Department Average Readiness Score")
print(df.groupby("department")["readiness_score"].mean())

print("\nTop Performer")
top = df.loc[df["readiness_score"].idxmax()]
print(top[["name","department","readiness_score"]])

print("\nStudents Needing Support")
print(df[df["Performance_Band"]=="Support"][["name","department","readiness_score"]])
```

 **Explanation**

In this step, I found the average readiness score for each department. I also identified the top performer and displayed the students who need support based on their performance band.
