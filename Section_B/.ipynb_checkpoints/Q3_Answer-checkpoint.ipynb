Q3. Variables, Memory and Identity5
Explain what happens in memory for the following code. Write expected relationship between values and object identity.

a = 25

b = a

c = 25

names = ["Asha", "Riya"]

same_names = names

same_names.append("Neha")


Answer:

When a = 25 is executed  python creates the value 25 in memory.
When b = a is executed it points to the same object as a.
When c = 25  is executed it points to the same object as a and b.
names = ["Asha", "Riya"]vit creates a list in memory.
same_names = names it points to the same list.
After same_names.append("Neha"), this list becomes ["Asha", "Riya", "Neha"].
Both names and same_names refer to the same list.
